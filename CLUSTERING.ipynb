```
from sklearn.cluster import KMeans
```

```
K_range = range(1,15)
wcss = []
```

```
for k in K_range:
    km = KMeans(n_clusters=k)
    km.fit(df[['wine','Magnesium']])
    wcss.append(km.inertia_)  #Inertia_ is the sum of squared error for each cluster.
```

```
wcss
```

```
plt.xlabel('Number of clusters (K)')
plt.ylabel('Sum of squared error')
plt.plot(K_range,wcss)
plt.title('WINE')
```

```
K_range = range(1,30)
wcss2 = []
```

```
for k in K_range:
    km = KMeans(n_clusters=k)
    km.fit(df[['Nonflavanoid phenols']])
    wcss2.append(km.inertia_)
```

```
wcss2
```

```
plt.xlabel('Number of clusters (K)')
plt.ylabel('Sum of squared error')
plt.plot(K_range,wcss2)
plt.title('Nonflavanoid phenols')
```
